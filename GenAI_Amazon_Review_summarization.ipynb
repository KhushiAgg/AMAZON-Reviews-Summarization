```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/OnePlus-Nord-Pastel-128GB-Storage/product-reviews/B0BY8JZ22K/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
code = requests.get(url)
code
```

```python
soup = BeautifulSoup(code.content,'html.parser')
```

```python
names = soup.select('span.a-profile-name')[2:]
names
```

```python
len(names)
```

```python
stars = soup.select('i.review-rating span.a-icon-alt')[2:]
stars
```

```python
len(stars)
```

```python
titles = soup.select('a.review-title span')
titles
```

```python
len(titles)
```

```python
dates = soup.select('span.review-date')[2:]
dates
```

```python
len(dates)
```

```python
reviews = soup.select('span.review-text span')
reviews
```

```python
len(reviews)
```

```python
names
```

```python
names[0].get_text()
```

```python
customer_name = []
ratings = []
rev_date = []
rev_title = []
rev_content = []
for i in range(10):
  customer_name.append(names[i].get_text())
  ratings.append(stars[i].get_text().replace(" out of 5 stars",""))
  rev_date.append(dates[i].get_text().replace("Reviewed in India on ",""))
  rev_title.append(titles[i].get_text())
  rev_content.append(reviews[i].get_text().strip("\n "))
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = rev_date
df['Customer Name'] = customer_name
df['Ratings'] = ratings
df['Title'] = rev_title
df['Reviews'] = rev_content
df
```

```python
df.to_csv("reviews.csv")
```

```python
import torch
from transformers import pipeline

summarizer = pipeline("summarization")
```

```python
for rev in rev_content:
  print(rev)
  summary = summarizer(rev, max_length=50, min_length=30, do_sample=False)
  print(summary)
```
